# Histograms

```python
data.DataReader(name="GOOG", data_source="google", start="2004-01-01", end="2016-12-31")
## ## THIS PULL DOES NOT WORK, SO I CANNOT TEST THIS CODE
```

```python
#custom function to create a buckets
def custom_round(sandwich_price):
    return int(sandwich_price / 1.0) * 1
```

```python
sandwich = pd.read_csv('bigmac.csv', index_col="Date")
sandwich.head(3)
```

```python
custom_round(123.32)  # it will round down to closest value of 1
```

```python
sandwich["Price in US Dollars"].apply(custom_round).value_counts().sort_index()
```

```python
sandwich["Price in US Dollars"].apply(custom_round).nunique()
```

```python
sandwich["Price in US Dollars"].apply(custom_round).plot(kind="hist", bins = 10)
```
