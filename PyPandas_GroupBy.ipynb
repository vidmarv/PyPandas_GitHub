# Methods on the Groupby Object and DataFrame Columns

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors.get_group("Apparel")["Revenue"].sum() #we are double checking if revenue in Apparel sector is OK
sectors.get_group("Apparel")["Profits"].sum()
```

```python
sectors.max()  # it displays company that is last in that sector alphabetically
sectors.min()  # it displays company that is first in that sector alphabetically
sectors.sum() #applies only to columns that have numeric numbers
```

```python
sectors["Revenue"].sum()  #call on individual column
sectors["Employees"].sum()
sectors["Profits"].max()
sectors["Profits"].min()
sectors["Employees"].mean()

sectors[["Revenue", "Profits"]].sum()
```

# The .agg() Method

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors.mean()
sectors["Employees"].mean()
```

```python
#agg method allows me to specify column and what method I want to use
# agg method expects Python dictionary {key:value} == {column_name : name_of_agg_method}
sectors.agg({"Revenue" : "sum",
            "Profits":"sum",
            "Employees" : "mean"})
```

```python
#I want to apply multiple operations to multiple columns
sectors.agg(["size", "sum", "mean"]) 
#we have taken sum and mean of every sector by grouping and calling agg method
```

```python
#combination of both methods
sectors.agg({"Revenue" : ["sum","mean"],
            "Profits":"sum",
            "Employees" : "mean"})
```

# Iterating through Groups

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors["Profits"].max()
```

```python
#new df object (empty) will be constructed and 
#column names will be the same as in fortune.csv
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
#2 temporary variables must be declared - sector and data
#data is one of 21 one DataFrames
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby("Location")
cities
#we override old DataFrame
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_in_city)
df
```
